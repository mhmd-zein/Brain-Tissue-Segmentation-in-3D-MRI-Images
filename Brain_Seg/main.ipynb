```python
import monai
import torch
from Training.Engine2D import Engine
from utils import set_seed
from monai.networks.nets import UNet, BasicUNet,SegResNet
from monai.losses import DiceLoss, DiceFocalLoss,DiceCELoss
from torch.optim.lr_scheduler import StepLR, CosineAnnealingWarmRestarts
from Dataset.Loaders2_5D import get_loader2_5D
from Dataset.Transforms2D import transforms2D,transforms2D_val
import numpy as np

data_path = r"/content/drive/MyDrive/Dataset001_BrainTissues"
train_loader = get_loader2_5D(
    data_path,
    transforms=transforms2D,
    shuffle=True,
    batch_size=32
)[0]

val_loader = get_loader2_5D(
    data_path,
    transforms=transforms2D_val,
    shuffle=False,
    batch_size=1
)[1]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# network = BasicUNet(
#     spatial_dims=2,
#     in_channels=3,
#     out_channels=4,
#     features=(32, 32, 64, 128, 256, 32),
#     dropout=0
# ).to(device)

network = UNet(
    spatial_dims=2,
    in_channels=3,
    out_channels=4,
    channels=(64, 128, 256, 512),
    strides=(2, 2, 2),
    num_res_units=3,
    norm='batch',
    bias=False,
    dropout=0.1,
).to(device)


max_epochs=120
optimizer = torch.optim.Adam(network.parameters(), lr=0.001)
scheduler, step = CosineAnnealingWarmRestarts(optimizer, T_0=60 * len(train_loader)), 'batch'
loss = DiceFocalLoss(softmax=True, jaccard=True, include_background=False, batch=True).to(device)

checkpoint_path = f"/content/drive/MyDrive/Brain_tissue_segmentation/Brain_Seg/best_checkpoint_basic_2D_73.pth"
checkpoint = torch.load(checkpoint_path, map_location=device)

network.load_state_dict(checkpoint['model_state_dict'])
optimizer.load_state_dict(checkpoint['optimizer_state_dict'])

for state in optimizer.state.values():
    for k, v in state.items():
        if isinstance(v, torch.Tensor):
            state[k] = v.to(device)

start_epoch = checkpoint['epoch']
best_Dice_score = checkpoint['best_Dice_score']
results = checkpoint['results']

print(f"Checkpoint loaded: Starting from epoch {start_epoch} with best Dice score: {best_Dice_score}")

engine = Engine(network, loss, optimizer, train_loader, val_loader, scheduler, step)
```
